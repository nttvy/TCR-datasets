```r
library(data.table)
library(conveniencefunctions)
library(VennDiagram)
library(tidyverse)
```

# Load data and check for exceptions

```r
dat <- fread('IEDB-160622.csv', data.table = F)
dim(dat)
head(dat)
```

```r
colnames(dat)
```

```r
dat <- dat[, !grepl('Chain 1', colnames(dat)) & !grepl('CDR1|CDR2', colnames(dat))]
dat <- dat[, c('Description', 'Antigen', 'Organism', 'Response Type', 'MHC Allele Names', 
               'Chain 2 CDR3 Curated', 'Chain 2 CDR3 Calculated')]
```

```r
head(dat)
```

```r
any(is.na(dat))
```

```r
dat[dat$`Chain 2 CDR3 Curated` != '' & dat$`Chain 2 CDR3 Curated` != dat$`Chain 2 CDR3 Calculated`,]
```

```r
dat$`Chain 2 CDR3 Curated` <- unlist(lapply(dat$`Chain 2 CDR3 Curated`, function(x) 
    if(grepl('^C', x) & grepl('F$', x)) {
        gsub('^C', '', x) %>% gsub('F$', '', .)
        } else {x}))

dat$`Chain 2 CDR3 Calculated` <- unlist(lapply(dat$`Chain 2 CDR3 Calculated`, function(x) 
    if(grepl('^C', x) & grepl('F$', x)) {
        gsub('^C', '', x) %>% gsub('F$', '', .)
        } else {x}))
```

```r
dat[dat$`Chain 2 CDR3 Curated` != '' & dat$`Chain 2 CDR3 Calculated` != '' & dat$`Chain 2 CDR3 Curated` != dat$`Chain 2 CDR3 Calculated`,]
```

```r
bool_exact <- dat$`Chain 2 CDR3 Curated` == dat$`Chain 2 CDR3 Calculated` | dat$`Chain 2 CDR3 Curated` == '' | dat$`Chain 2 CDR3 Calculated` == ''
bool_curated <- (gsub('^C', '', dat$`Chain 2 CDR3 Curated`) == dat$`Chain 2 CDR3 Calculated`) | 
                (gsub('F$', '', dat$`Chain 2 CDR3 Curated`) == dat$`Chain 2 CDR3 Calculated`)
bool_calculated <- (gsub('^C', '', dat$`Chain 2 CDR3 Calculated`) == dat$`Chain 2 CDR3 Curated`) | 
                   (gsub('F$', '', dat$`Chain 2 CDR3 Calculated`) == dat$`Chain 2 CDR3 Curated`)
bool <- bool_exact | bool_curated | bool_calculated
table(bool_exact)
table(bool_curated)
table(bool_calculated)
table(bool)
```

```r
dat[!bool,] # to be filtered
```

```r
dat <- dat[bool,]
dim(dat)
```

```r
dat <- dat[!(dat$`Chain 2 CDR3 Curated` == '' & dat$`Chain 2 CDR3 Calculated` == ''),]
dim(dat)
```

```r
dat[dat$`Chain 2 CDR3 Curated` == dat$`Chain 2 CDR3 Calculated`,]
```

```r
dat$CDR3b <- ifelse(dat$`Chain 2 CDR3 Curated` != '' & length(dat$`Chain 2 CDR3 Curated`) <= length(dat$`Chain 2 CDR3 Calculated`), 
                    dat$`Chain 2 CDR3 Curated`, dat$`Chain 2 CDR3 Calculated`)
```

```r
# check for special characters in CDR3 and epitope sequences
any(grepl("[[:punct:]]", dat$CDR3b))
any(grepl("[[:punct:]]", dat$Description))
```

```r
dat$Description[grepl("[[:punct:]]", dat$Description)]
```

```r
dat <- dat[!grepl("[[:punct:]]", dat$Description),]
dim(dat)
```

```r
# check for special characters in CDR3 and epitope sequences
any(grepl(" ", dat$CDR3b))
any(grepl(" ", dat$Description))
```

```r
dat$Description[grepl(" ", dat$Description)]
```

```r
dat <- dat[!grepl(" ", dat$Description),]
dim(dat)
```

```r
head(dat)
```

```r
dat <- dat[!grepl('[a-z]', dat$Description),] # remove lower case 
dat <- dat[!grepl('[a-z]', dat$CDR3b),]
dim(dat)
```

```r
dat <- dat[!grepl('[0-9]', dat$Description),] # remove digits
dim(dat)
```

```r
dat <- unique(dat)
dim(dat)
```

```r
# write loaded data
write.table(dat, 'IEDB-loaded.csv', quote = F, row.names = F, sep = ',', append = F)
```

```r
length(unique(dat$`MHC Allele Names`))
sort(unique(dat$`MHC Allele Names`))
```

```r
sort(table(dat$`MHC Allele Names`))
```

```r
unclear_hla <- unique(dat$`MHC Allele Names`)[grepl('[,/]', unique(dat$`MHC Allele Names`)) 
                                              | grepl('class', unique(dat$`MHC Allele Names`))]
length(unclear_hla)
unclear_hla
```

```r
nrow(dat[dat$`MHC Allele Names` %in% unclear_hla,])
nrow(dat)
sort(table(dat[dat$`MHC Allele Names` %in% unclear_hla, 'MHC Allele Names']), decreasing = TRUE)
```

```r
dat <- dat[dat$`MHC Allele Names` != '' & !dat$`MHC Allele Names` %in% unclear_hla,]
table(dat$`MHC Allele Names`)
dim(dat)
```

```r
length(dat[!grepl('[*]', dat$`MHC Allele Names`), 'MHC Allele Names'])
table(dat[!grepl('[*]', dat$`MHC Allele Names`), 'MHC Allele Names'])
```

```r
old_hla <- c('HLA-A1', 'HLA-A11', 'HLA-A2', 'HLA-B18', 'HLA-B27', 'HLA-B35', 'HLA-B57', 'HLA-B7', 'HLA-B8', 
             'HLA-Cw3', 'HLA-DPw2')
new_hla <- c('HLA-A*01', 'HLA-A*11', 'HLA-A*02', 'HLA-B*18', 'HLA-B*27', 'HLA-B*35', 'HLA-B*57', 'HLA-B*07', 
             'HLA-B*08', 'HLA-C*03:04', 'HLA-DP2')
same_hla <- unique(dat$`MHC Allele Names`)[!unique(dat$`MHC Allele Names`) %in% old_hla]
old_hla <- c(old_hla, same_hla)
new_hla <- c(new_hla, same_hla)
dat$`MHC Allele Names` <- plyr::mapvalues(dat$`MHC Allele Names`, from = old_hla, to = new_hla)
unique(dat$`MHC Allele Names`)
```

```r
dat <- unique(dat[, c('Description', 'MHC Allele Names', 'CDR3b')])
dim(dat)
```

```r
write.table(dat, 'IEDB-analyzed.csv', quote = F, append = F, row.names = F, sep = ',')
```
